```python
import seaborn as sns
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
sns.set()

FILE = "../../data/capture-850-5.csv"
BASELINE = "../../data/capture-baseline.csv"
SERVER_PORTS = [443, 80]
```

```python
data = (pd.read_csv(FILE, header=None, names=["time", "length"]).sort_values(by="time"))
baseline = (pd.read_csv(BASELINE, header=None, names=["time", "length"]).sort_values(by="time"))

display(data, baseline)
```

```python
fig, (axes, baseline_axes) = plt.subplots(2, 1, sharex=True, sharey=True, figsize=(8,6))

sns.scatterplot(x=data["time"], y=data["length"], marker='x', ax=axes, edgecolor="none")
sns.scatterplot(x=baseline["time"], y=baseline["length"], marker='x', ax=baseline_axes, edgecolor="none")

axes.set_xlim(0, 2)
axes.set_ylim(-2000, 2000)

# fig.savefig("vanilla-template-shared.pdf", dpi=300)
```
